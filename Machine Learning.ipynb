## Standardize the Data
to subtract the mean and devide by the standard deviation featurewise (DONT INCLUDE THE DUMMIES / ONE HOT ENCODED)

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
#THIS CLASS ALLOW TO CHOOSE WHAT FEATURE TO BE TRANSFORM
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator,TransformerMixin):
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns=columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self,X,y=None):
        self.scaler.fit(X[self.columns],y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self,X,y=None,copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled],axis=1)[init_col_order]
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature name'],data = feature_name)

summary_table['coefficient'] = np.transpose(reg.coef_)

summary_table
```

```python
#adding intercept
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
```

```python
#log(odds) = intercept + b0 + b1x1 + b2x2 + b3x3 + ... + bnxn
#where coefficient is weight, intercept is bias
#the closer the coefficient to 0 = smaller weight
summary_table
```

## Interpreting the Coefficient

```python
summary_table['Odd Ratio'] = np.exp(summary_table.coefficient)
summary_table
```

```python
summary_table.sort_values('Odd Ratio',ascending=False)
```

### NOTES
A feature is not particularly important: <br>
    if its coefficient is around 0<br>
    if its odds ratio is around 1<br>

A weight (coefficient) of 0 implies that no matter the feature value, we will multiply it by 0 (in the model) <br>

For a unit change in the standardized feature, the odds increase by a multiple equal to the odds ratio (1 = no change) <br>
<i>*which are month, distance to work, daily work load average, and day of the week is not a good feature</i>
